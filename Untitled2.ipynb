```python
state = pd.read_csv('/Users/jparkgeo/Git_Repo/agingdam/dam_risk/tiger_tract_2020/state_lookup.csv')
state = state.loc[state['ContiguousUS'] == 1].reset_index(drop=True)
state['FIPS'] = state['FIPS'].astype(str)
state['FIPS'] = state.apply(lambda x: '0'+x['FIPS'] if len(x['FIPS']) == 1 else x['FIPS'], axis=1)
state
```

```python
for state_num in state['FIPS']:
    block = gpd.read_file(f'/Users/jparkgeo/Git_Repo/agingdam/dam_risk/tiger_block_2020/tl_2020_{state_num}_tabblock20')
    block = block[['GEOID20', 'HOUSING20', 'POP20', 'geometry']]
    block.rename(columns={'GEOID20': 'GEOID'}, inplace=True) 
    block = block.to_crs(epsg=4326)
    block.to_file(f'/Users/jparkgeo/Git_Repo/population_vulnerable_to_dam_failure/census_geometry/tl_2020_{state_num}_tabblock20.geojson')
    print(state_num, block.crs.name)
```
