```python
import geopandas as gpd
import os
import matplotlib.pyplot as plt
```

Change the following cell to the directory which stores output files. 

```python
path = './output'
```

The following cell will plot polygonized inundated map for every dam.

```python
fim = gpd.read_file(os.path.join(path, 'TAS_F_fim.geojson'))
fim.plot('value', figsize=(30, 30), cmap='Blues')
```

The following cell will plot each dam with inundated map and its ellipse. 

```python
ellipse = gpd.read_file(os.path.join(path, 'TAS_F_ellipse.geojson'))

for dam_id in ellipse['Dam_ID']:
    fig, ax = plt.subplots(1,1, figsize=(10,10))
    print(dam_id)
    fim.loc[fim['Dam_ID'] == dam_id].plot('value', ax=ax, cmap='Blues')
    ellipse.loc[ellipse['Dam_ID'] == dam_id].boundary.plot(ax=ax)
    plt.show()
```

The following cell will plot census block and ellipse. You can change the file name (`TAS_F_lm.geojson`) according to your file, indicating `fim_geoid_gdf`.

```python
lm = gpd.read_file(os.path.join(path, 'TAS_F_lm.geojson'))

for dam_id in lm['Dam_ID'].unique():
    fig, ax = plt.subplots(1,1, figsize=(10,10))
    print(dam_id)
    lm.loc[lm['Dam_ID'] == dam_id].plot('Class', ax=ax, cmap='Blues')
    ellipse.loc[ellipse['Dam_ID'] == dam_id].boundary.plot(ax=ax)
    plt.show()
```
